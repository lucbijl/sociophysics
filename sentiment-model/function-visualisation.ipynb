```python
import numpy as np
import matplotlib.pyplot as plt

def age_function(var, params):
    return np.exp(-(params * var))

# Define the domain and range
days = np.linspace(0, 50, 1000)  # 100 points from 0 to 10 days
params = 0.1  # You can adjust this value based on your needs

# Calculate the function values for the given domain and parameters
values = age_function(days, params)

# Plot the function
plt.figure(figsize=(10,6))
plt.plot(days, values, color='tab:red')
plt.title(r'Age function $g_{n+1}$ with $\mu = 0.1$')
plt.xlabel(r'$p_{n+1}$ in days')
plt.xlim(0,50)
plt.ylabel(r'Probability $g_{n+1}$')
plt.ylim(0, 1.4)
plt.grid(True)
plt.tick_params(axis='both', direction='in',top=True, right=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the sum of sigmoid functions.
def logistic_sum(var, params):
    output = 0
    for a, b, c in params:
        output += a / (1 + np.exp(-b * (var - c)))
    return np.clip(output, 0, 1)

# Parameters for the sigmoid_sum function
parameters_dict = {
    'sentiment': [[0.4, 1.5, 1], [0.6, -1.5, -1]],
    'valence': [[0.4, 1.5, 1], [0.6, -1.5, -1]],
    'arousal': [[1, 2, 0]],
    'dominance': [[1, 0.5, 0]]
}

# Create a 2 by 2 subplot
fig, axes = plt.subplots(2, 2, figsize=(10, 9))

# Define the domain and range
var_domain = np.linspace(-1, 1, 100)
var_range = np.linspace(0, 1, 100)

# Iterate through parameters and plot on each subplot
for ax, (key, params), i in zip(axes.flat, parameters_dict.items(), [1,2,3,4]):
    # Calculate the function values for the given domain and parameters
    values = logistic_sum(var_domain, params)

    # Plot the function
    ax.plot(var_domain, values, color='tab:red')
    ax.set_title(fr'Logistic {key} subset probability $g_{i}$')
    ax.set_xlabel(fr'$p_{i}$ {key}')
    ax.set_xlim(-1,1)
    ax.set_xticks(np.linspace(-1, 1, 6))
    ax.set_ylabel(fr'Probability $g_{i}$')
    ax.set_ylim(0, 1)
    ax.grid(True)
    ax.tick_params(axis='both', direction='in',top=True, right=True)

# Adjust layout to prevent clipping of titles
plt.tight_layout()
plt.show()
```
